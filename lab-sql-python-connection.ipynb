```python
def rentals_month(engine, month, year):

    query = text("""
    SELECT *
    FROM rental
    WHERE EXTRACT(MONTH FROM rental_date) = :month
      AND EXTRACT(YEAR FROM rental_date) = :year
    """)
    
    with engine.connect() as connection:
        df = pd.read_sql(query, connection, params={"month": month, "year": year})
    
    return df

user = "root"
pw = getpass()
db = "sakila"

engine = create_engine(f"mysql+pymysql://{user}:{pw}@localhost/{db}")

month = int(input("Input the month number: "))
year = int(input("Input the year: "))
df_rentals = rentals_month(engine, month, year)

df_rentals
```
